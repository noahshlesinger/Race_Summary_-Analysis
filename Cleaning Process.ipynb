```python
# Import dependencies
import pandas as pd
```

```python
# Read in Race Data
pd.read_csv("race_data.csv")
race_df = pd.read_csv("race_data.csv")
race_df
```

```python
# Only include participants with finish times
finishers_df = race_df.dropna()
finishers_df.head(15)
```

```python
# Creates a funtion to convert time from H:M:S to seconds as an integer value

def second_conversion(time):
    t = str(time).split(":")
    h = int(t[0])
    m = int(t[1])
    s = int(t[2])
    return (h*60*60) + (m*60) + s            
```

```python
# Converst Swim split, bike split, run split, and total time to number of seconds
finishers_df["SWIM"] = finishers_df["SWIM"].map(second_conversion)
finishers_df["BIKE"] = finishers_df["BIKE"].map(second_conversion)
finishers_df["RUN"] = finishers_df["RUN"].map(second_conversion)
finishers_df["TOTAL TIME"] = finishers_df["TOTAL TIME"].map(second_conversion)
```

```python
# Remove decimal point and check finishers_df for accurate conversion
finishers_df = finishers_df.astype({'BIB':'int'})
finishers_df
```
